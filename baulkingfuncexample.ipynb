```python
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import linspace
import seaborn as sns
sns.set_style('whitegrid')
```

```python
def b(n, beta):
    if n == 0:
        return 0
    else:
        return 1 - (beta/n)
```

```python
b025 = [b(i, 0.25) for i in range(10)]
b05 = [b(i, 0.5) for i in range(10)]
b075 = [b(i, 0.75) for i in range(10)]
b099 = [b(i, 1.00) for i in range(10)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
colors = [ cm.plasma(x) for x in linspace(0.15, 0.85, 4) ]
plt.plot(range(10), b025, c=colors[0], label=r'$\beta = 0.25$')
plt.plot(range(10), b05, c=colors[1], label=r'$\beta = 0.50$')
plt.plot(range(10), b075, c=colors[2], label=r'$\beta = 0.75$')
plt.plot(range(10), b099, c=colors[3], label=r'$\beta = 1.00$')
plt.legend(loc=4, fontsize=16)
plt.xlabel('$m$', fontsize=16)
plt.ylabel('$b_2(m)$', fontsize=16)
plt.title(r'Examples of the Baulking Function $b_2(m)$', fontsize=18)
plt.savefig('tex/img/examplebaulking2.pdf')
```

```python
def b1(n, N):
    return n / (N+1)
```

```python
b2 = [b1(i, 2) if i <= 3 else None for i in range(10)]
b4 = [b1(i, 4) if i <= 5 else None for i in range(10)]
b6 = [b1(i, 6) if i <= 7 else None for i in range(10)]
b8 = [b1(i, 8) if i <= 9 else None for i in range(10)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
colors = [ cm.plasma(x) for x in linspace(0.15, 0.85, 4) ]
plt.plot(range(10), b2, c=colors[0], label=r'$n = 2$')
plt.plot(range(10), b4, c=colors[1], label=r'$n = 4$')
plt.plot(range(10), b6, c=colors[2], label=r'$n = 6$')
plt.plot(range(10), b8, c=colors[3], label=r'$n = 8$')
plt.legend(loc=4, fontsize=16)
plt.xlabel('$m$', fontsize=16)
plt.ylabel('$b_1(m)$', fontsize=16)
plt.title(r'Examples of the Baulking Function $b_1(m)$', fontsize=18)
plt.savefig('tex/img/examplebaulking1.pdf')
```
