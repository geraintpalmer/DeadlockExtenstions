```python
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import linspace
import seaborn as sns
sns.set_style('whitegrid')
```

```python
def b(n, beta):
    if n == 0:
        return 0
    else:
        return 1 - (beta/n)
```

```python
b025 = [b(i, 0.25) for i in range(10)]
b05 = [b(i, 0.5) for i in range(10)]
b075 = [b(i, 0.75) for i in range(10)]
b099 = [b(i, 0.99) for i in range(10)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
colors = [ cm.plasma(x) for x in linspace(0.15, 0.85, 4) ]
plt.plot(range(10), b025, c=colors[0], label=r'$\beta = 0.25$')
plt.plot(range(10), b05, c=colors[1], label=r'$\beta = 0.50$')
plt.plot(range(10), b075, c=colors[2], label=r'$\beta = 0.75$')
plt.plot(range(10), b099, c=colors[3], label=r'$\beta = 0.99$')
plt.legend(loc=4, fontsize=16)
plt.xlabel('$x$', fontsize=16)
plt.ylabel('$b(x)$', fontsize=16)
plt.title(r'Examples of the Baulking Function $b(x)$', fontsize=18)
plt.savefig('tex/img/examplebaulking.pdf')
```
